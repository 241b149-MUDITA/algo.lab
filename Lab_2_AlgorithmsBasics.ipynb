```python
def func8(A): # Renamed from func8 to avoid confusion with the helper, but the task specifies func8
    n = len(A)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
    return A
```

### Explanation of `func8` (Bubble Sort Implementation)

The `func8` function now implements the **Bubble Sort** algorithm. This algorithm sorts an array by repeatedly stepping through the list, comparing adjacent elements and swapping them if they are in the wrong order. The pass through the list is repeated until no swaps are needed, which indicates that the list is sorted.

**How it works:**

1.  **Outer Loop (`for i in range(n)`):** This loop iterates `n` times, where `n` is the length of the array. In each iteration, the largest unsorted element "bubbles" to its correct position at the end of the unsorted portion of the array. After `i` iterations, the last `i` elements of the array are sorted and in their final positions.

2.  **Inner Loop (`for j in range(0, n - i - 1)`):** This loop performs the comparisons and swaps for the current pass. It iterates from the beginning of the array up to the last unsorted element (`n - i - 1`).

3.  **Comparison and Swap (`if A[j] > A[j + 1]`):** Inside the inner loop, adjacent elements `A[j]` and `A[j + 1]` are compared. If `A[j]` is greater than `A[j + 1]`, they are in the wrong order (for ascending sort), so they are swapped.

**Time Complexity:**

*   **Worst-case and Average-case:** O(n^2), where `n` is the number of elements in the array. This occurs when the array is in reverse order or elements are randomly distributed, as each element might need to be compared and swapped multiple times.
*   **Best-case:** O(n). This occurs when the array is already sorted, and the algorithm can detect this by not performing any swaps in the first pass (though the current implementation doesn't include an early exit optimization, so it would still perform O(n^2) comparisons even for an already sorted array).

**Space Complexity:**

*   O(1), as it only uses a constant amount of extra space for temporary variables during swaps, regardless of the input array size.
